```python
import matplotlib.pyplot as plt
import numpy
data=numpy.loadtxt(fname = "inflammation-01.csv", delimiter=',')
numpy.max(data, axis=1)
numpy.min(data,axis=1)
numpy.mean(data, axis=0)

ave_inflammation = numpy.mean(data, axis=0)
ave_plot = plt.plot(ave_inflammation)
```

```python
fig=plt.figure(figsize=(10.0,3.0))
a =fig.add_subplot(3,1,1)
b =fig.add_subplot(3,1,2)
c =fig.add_subplot(3,1,3)
a.set_ylabel("average")
a.plot(numpy.mean(data, axis=0))
b.set_ylabel("min")
b.plot(numpy.min(data,axis=0))
c.set_ylabel("max")
c.plot(numpy.max(data,axis=0))
```

```python
fig=plt.figure(figsize=(10.0,3.0))
a =fig.add_subplot(1,3,1)
b =fig.add_subplot(1,3,2)
c =fig.add_subplot(1,3,3)
a.set_xlabel("average")
a.plot(numpy.mean(data, axis=0))
b.set_xlabel("min")
b.plot(numpy.min(data,axis=0))
c.set_xlabel("max")
c.plot(numpy.max(data,axis=0))
fig.tight_layout()
plt.savefig("inflammation.png")
plt.show()
```
